```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Data Cleaning**

```python
# uploading dataset
main_df = pd.read_csv("/content/Sales_Data_Analysis_Visualization.csv", encoding="unicode_escape")
main_df.head(10)
```

```python
# finding the shape of the dataset
main_df.shape
```

```python
# examining the necessary information about the dataset like datatypes and missing values etc.
main_df.info()
```

```python
# deleting the status and unnamed colunm because it contains all the missing(null) values and do not need to be in dataset for usage
main_df.drop(["Status", "unnamed1"], inplace=True, axis=1)
```

```python
# further checking the missing(null) values
main_df.isnull().sum()
```

```python
# deleting/removing all the missing(null) values from the columns
main_df.dropna(inplace=True)
```

```python
# checking missing(null) values
main_df.isnull().sum()
```

```python
# changing the datatype of the Amount column from float into integer
main_df["Amount"] = main_df["Amount"].astype("int")
```

```python
# test
main_df.info()
```

```python
# describing mean, median, mode, min, max, and percentiles of the dataset
main_df.describe()
```

# **Exploratory Data Analysis**

# **Gender Data Analysis**

```python
# visualize the number of males and females in the dataset
ax = sns.countplot(x="Gender", data=main_df)

# running a loop to show values on the bars
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
# finding the total amount spent by males and females
gender_output_sales = main_df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent by males and females
ax = sns.barplot(x="Gender", y="Amount", data=gender_output_sales)
```

**Gender: From above graphs we can see that most of the buyers are females and even the purchasing power of females is greater than males**

# **Age Group Data Analysis**

```python
# Visualizing the males and females according to their age
ax = sns.countplot(x="Age Group", hue="Gender", data=main_df)

# loop for placing values on the bars
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
# finding the total amount spent according to age
age_output_sales = main_df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent according to age
ax = sns.barplot(x="Age Group", y="Amount", data=age_output_sales)
```

**Age: From above graphs we can see that most of the buyers are of age group between 26-35 years which are females**

# **State Data Analysis**

```python
# finding the total orders according to state
state_output_orders = main_df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(15)

# Visualization of the total orders according to state
sns.set(rc={"figure.figsize":(13,5)})
sns.barplot(x="State", y="Orders", data=state_output_orders)
plt.xticks(rotation="vertical")
plt.show()
```

```python
# finding the total orders according to state
state_output_sales = main_df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(15)

# Visualization of the total orders according to state
sns.set(rc={"figure.figsize":(13,5)})
sns.barplot(x="State", y="Amount", data=state_output_sales)
plt.xticks(rotation="vertical")
plt.show()
```

**State: From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively**

# **Marital Status Data Analysis**

```python
# Visualizing the males and females according to their age
ax = sns.countplot(x="Marital_Status", data=main_df)
sns.set(rc={"figure.figsize":(5,3)})

# loop for placing values on the bars
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
# finding the total amount spent according to age
mar_output_sales = main_df.groupby(["Marital_Status"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent according to age
ax = sns.barplot(x="Marital_Status", y="Amount", data=mar_output_sales)
```

```python
# finding the total amount spent according to marital status
mar_vis_output_sales = main_df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent according marital status
ax = sns.barplot(x="Marital_Status", y="Amount", hue="Gender", data=mar_vis_output_sales)
sns.set(rc={"figure.figsize":(15,10)})
```

**Marital Status: From above graphs we can see that most of the buyers are married and are mostly women and they have high purchasing power**

# **Occupation Data Analysis**

```python
# Visualizing the total occupations
ax = sns.countplot(x="Occupation", data=main_df)
sns.set(rc={"figure.figsize":(25,12)})

# loop for placing values on the bars
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
occup_output_sales = main_df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent according marital status
ax = sns.barplot(x="Occupation", y="Amount", data=occup_output_sales)
sns.set(rc={"figure.figsize":(15,10)})
```

**Occupation: From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector**

# **Product Category Data Analysis**

```python
# Visualizing the total sales according to product categories
ax = sns.countplot(x="Product_Category", data=main_df)
sns.set(rc={"figure.figsize":(25,12)})
plt.xticks(rotation="vertical")

# loop for placing values on the bars
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
prod_output_sales = main_df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

# Visualization of the total amount spent according marital status
ax = sns.barplot(x="Product_Category", y="Amount", data=prod_output_sales)
sns.set(rc={"figure.figsize":(15,10)})
plt.xticks(rotation="vertical")
plt.show()
```

**Product Category: From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category**

# **Product ID Data Analysis**

```python
# Examining the most sold products
proID_output_sales = main_df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)

# Visualization of the most sold products
ax = sns.barplot(x="Product_ID", y="Orders", data=proID_output_sales)
sns.set(rc={"figure.figsize":(8,5)})
plt.xticks(rotation="vertical")
plt.show()
```
